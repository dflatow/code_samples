```python
import pandas as pd
import numpy as np
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
```

```python
def compare_size(new, old):
    frac_removed = 1.0 - 1.0 * len(new) / len(old)
    print "%0.2f percent removed" % (100.0 * frac_removed, )
```

```python
df_orig = pd.read_csv('../data/wine.train.csv')
#df = df[df.color == 'white']
df_orig = df_orig.drop(['Unnamed: 0', 'color'], axis=1)
df = df_orig.copy()

df = df[df.density < 1.005]
compare_size(df, df_orig)

df = df[df['free.sulfur.dioxide'] < 200]
compare_size(df, df_orig)

df = df[df['chlorides'] < 0.30]
compare_size(df, df_orig)

df = df[df['citric.acid'] < 1.0]
compare_size(df, df_orig)
```
