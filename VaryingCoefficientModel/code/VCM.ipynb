```python
import pandas as pd
```

```python
df = pd.read_csv('../data/vcdata.csv')
df = df.drop('Unnamed: 0', axis=1)
```

```python
bins = 15
df['period'] = pd.cut(df.t, bins, labels=False)
groups = df.groupby('period')
betas = []
for i, p in groups:
    m = pd.ols(y=p.y, x=p[['x1', 'x2']])
    betas.append(m.beta)
betas = pd.concat(betas, axis=1).T.plot()
```
